Q1: Explain the basic structure of an HTML page. Why is it important to follow
this structure?
Ans:- The below is used the basci structure of HTMl page
Each section is used is explained within comment

```
<!-- Specify which version of HTML is being used , here HTML 5 is used-->
<!DOCTYPE html>

<!-- html tag is ain wrapper of the page specify which language is beiing used here -->
<html lang="en">

<!-- head tag contains the meta data about the page like charset, page title , resources being used if any SEO keywords, viewport details etc. -->
  <head>
    <!-- meat tag used to specify which character encoding technique is used  -->
    <meta charset="UTF-8" />

    <!-- Title of the page -->
    <title>Title</title>
  </head>

 <!-- body tag contains the actual content which being rendered on the page -->
  <body>
    <!-- Content of the page -->
  </body>
</html>
```

Question 4: You’re creating a form that asks the user to enter their name, select their
country from a list, and click a button to submit. Write the HTML code for this.
Ans:

```
<form action="#" method="post">
  <!-- Name Input -->
  <input type="text" name="name" placeholder="Enter Name" required />

  <!-- Country Input with Datalist -->
  <input list="country" name="country" placeholder="Select or Type Country" required />

  <datalist id="country">
    <option value="Australia">
    <option value="United States">
    <option value="India">
    <option value="Russia">
  </datalist>

  <!-- Submit Button -->
  <input type="submit" value="Submit" />
</form>
```
